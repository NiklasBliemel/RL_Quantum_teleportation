```python
class ComplexTile:
    def __init__(self, grid_size, N_grids, left_bound=-1, right_bound=1):
        self.max_index = grid_size - 1
        self.N_grids = N_grids
        self.d = 1 / N_grids
        self.scale = (right_bound - left_bound)
        self.shift = -left_bound
        
    def __call__(self, x):
        out = nsphere_rep(x)
        n_state = len(out)
        out = (out + self.shift) / self.scale
        out = self.max_index * out.unsqueeze(0) + (self.d * torch.arange(self.N_grids).unsqueeze(-1))
        out = torch.trunc(out).int()
        out[out > self.max_index] = self.max_index
        out[out < 0] = 0
        return torch.matmul(out, (self.max_index + 1)**torch.arange(n_state).int())
```

```python
def nsphere_rep(state):
    new = torch.view_as_real_copy(state).flatten()
    psis = []
    for i in range(1,len(new)):
        if i == len(new)-1:
            psis.append(torch.atan2(new[i-1], new[i]) / torch.pi * 2)
        else:
            psis.append(torch.atan2(torch.sqrt(torch.sum(new[:i]**2)), new[i]) / torch.pi * 2)
    return torch.tensor(psis)
```
